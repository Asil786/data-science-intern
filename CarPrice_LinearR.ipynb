# Performing LinearRegression

```python
from sklearn.linear_model import LinearRegression
```

# Finding rmse and r^2

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
rmse = mean_squared_error(y_test,y_pred,squared=False) # by default squared=true i.e. mse,  for rmse squared=false
```
