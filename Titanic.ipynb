```python
df=df.drop(['PassengerId','Cabin','Ticket','Name','Embarked'],axis=1)
```

```python
test=test.drop(['PassengerId','Cabin','Ticket','Name','Embarked'],axis=1)
```

```python
x=df.drop('Survived',axis=1)
```

```python
y=df['Survived']
```

```python
sns.histplot(x['Age'])
```

```python
x.head()
```

```python
x['Age'].fillna(value=x['Age'].median(),inplace=True)
```

```python
test['Age'].fillna(value=test['Age'].median(),inplace=True)
```

```python
test.isnull().sum()
```

```python
sns.distplot(test['Fare'])
```

```python
test['Fare'].fillna(value=test['Fare'].median(),inplace=True)
```

```python
x['Sex']=x['Sex'].map({'male':1,'female':2})
```

```python
x.head()
```

```python
test['Sex']=test['Sex'].map({'male':1,'female':2})
```

```python
test.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=123,stratify=y)
```

KNN


```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(x_train,y_train)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logi=LogisticRegression()
```

```python
logi.fit(x_train,y_train)
```

```python
y_pred=logi.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
result=knn.predict(test)
```
