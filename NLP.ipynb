#Tokenization using Spacy

```python
import spacy
nlp=spacy.load('en_core_web_sm') #to load the language of British English
```

```python
doc=nlp("Wanna buy and looking house for $5 million")
```

```python
for t in doc:
  print(t.text)
```

```python
doc1=nlp("Wanna buy and looking house for $5 million")
for t in doc1:
  print (t.text,t.pos_)
```

```python
doc1=nlp('i am runner running in a race because i like to run since i ran today')
for i in doc1:
  print(i, '\t',i.lemma_)
```

```python
doc1=nlp('i am looking Screaming runner running in a race because i like to run since i ran today')
for i in doc1:
  print(i, '\t',i.lemma_)
```
