#VADER (Valence ware Dictionary for sEntiment Reasoning)

```python
nltk.download('vader_lexicon')  #for english language
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
sia=SentimentIntensityAnalyzer()
```

```python
a='This was a best movie' #score depends on the sentences or string values
sia.polarity_scores(a) #need the pass the value
```

```python
b='This was the best, most Awesome movie ever made!!!!!!'
sia.polarity_scores(b)
```

```python
c='This is ok ok movie'
sia.polarity_scores(c)
```

#Adding score to Dataframe

```python
df['Scores']=df['review'].apply(lambda r:sia.polarity_scores(r)) #apply the Algorithm for all the Statements
```

```python
df.head()
```

```python
df['compound']=df['Scores'].apply(lambda c:c['compound'])
```

```python
df.head()
```

#Creating pred_label column based on compound score

```python
df['pred_label']=df['compound'].apply(lambda m:'pos' if m>0 else 'neg')
```

```python
df.head()
```

#Finding Accuracy

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
Acc_sr=accuracy_score(df['label'],df['pred_label'])
Acc_sr
```

```python
print(classification_report(df['label'],df['pred_label']))
```

```python
confusion_matrix(df['label'],df['pred_label'])
```
