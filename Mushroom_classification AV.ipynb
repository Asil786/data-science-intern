```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(['class'],axis=1)
y = df['class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=123)
```

# Linear Regression model

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
```

```python
# fit function is to train

reg.fit(X_train,y_train)
```

```python
# give the value of c of the linear regression formula ( y = mx + c )
reg.intercept_
```

```python
# give the value of m of the linear regression formula ( y = mx + c )
reg.coef_
```

```python
y_pred = reg.predict(X_test)
y_pred
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
# dont relate squared with actual formula..
# if squared=True (defualt) -> SME value will be returned
# if squared=Flase -> RSME value will be returened
rmse = mean_squared_error(y_test,y_pred,squared=False)
rmse
```

```python
r2_score(y_test,y_pred)
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(X_train,y_train)
```

```python
y_pred = knn.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
logi=LogisticRegression()
```

```python
logi.fit(X_train,y_train)
```

```python
y_pred = logi.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
