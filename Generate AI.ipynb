```python
!pip install diffusers


!pip install transformers

from diffusers import StableDiffusionPipeline
import torch

model_id = "runwayml/stable-diffusion-v1-5"

sd_pipeline=StableDiffusionPipeline.from_pretrained(model_id,torch_dtype=torch.float16)

sd_pipeline = sd_pipeline.to ("cuda")  #cuda means set it to run on gpu

def get_completion_sd(prompt):
  negative_prompt = """
  simple background,duplicate,low qualtiy,lowest quality,bad anatomy,extra digits,lowres,username,artist name,error,duplicate,watermark,signature,text,blurry,worst qualtiy,bad properties"""

  return sd_pipeline(prompt,negative_prompt=negative_prompt).images[0]

prompt="Girl"

print(prompt)

sd_image=get_completion_sd(prompt)

sd_image.save("./horse.jpg")
```

```python
# Build AI Image generator

!pip install gradio==3.48.0

def get_completion_sd(prompt):
  negative_prompt = """
  simple background,duplicate,low qualtiy,lowest quality,bad anatomy,extra digits,lowres,username,artist name,error,duplicate,watermark,signature,text,blurry,worst qualtiy,bad properties"""

  return sd_pipeline(prompt,negative_prompt=negative_prompt).images[0]

import gradio as gr

genai_app = gr.Interface(fn=get_completion_sd,
    inputs=[gr.Textbox(label="Your Prompt")],
    outputs=[gr.Image(label="Result")],
    title="Generate Cool Images",
    description="Generate any image with stable diffusion model",
    allow_flagging="never",
    example=["Astronaut,riding a horse,on mars",
             "cargo ship,flying in space"])

genai_app.launch()
```
