```python
import pandas as pd
from sklearn import datasets
iris=datasets.load_iris()
```

```python
x=iris.data
y=iris.target
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=120,test_size=0.3,stratify=y)
```

```python
from xgboost import XGBClassifier
xgc=XGBClassifier(n_estimators=50)
```

```python
xgc.fit(x_train,y_train)
```

```python
y_pred=xgc.predict(x_test)
```

```python
from sklearn.metrics import f1_score,accuracy_score,confusion_matrix,classification_report
```

```python
print(accuracy_score(y_pred, y_test))
```

```python
print(confusion_matrix(y_pred, y_test))
```

```python
print(classification_report(y_pred, y_test))
```

# Hyper-Tuning

```python
parm={'learning_rate': [0.10, 0.05, 0.15,0.20,0.30],
    'max_depth': [3, 5, 7,6,8,10,12,15],
    'min_child_weight': [1, 3, 5, 7],
    'gamma': [0.0, 0.1, 0.2,0.3,0.4],
    'n_estimators': [40, 50, 100]}
```

```python
from sklearn.model_selection import GridSearchCV
gsv=GridSearchCV(estimator=xgc,param_grid= parm , scoring = 'accuracy')
```

```python
gsv.fit(x_train,y_train)
```

```python
gsv.best_params_
```

```python
gsv.best_score_
```

```python
xgc=XGBClassifier(gamma=0.0,
 learning_rate=0.1,
 max_depth=3,
 min_child_weight=1,
 n_estimators=40)
```

```python
xgc.fit(x_train,y_train)
```

```python
y_pred=xgc.predict(x_test)
```
